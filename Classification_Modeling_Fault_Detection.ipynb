# 1. 불량 데이터 읽기

```python
import pandas as pd
data = pd.read_csv("defect_train.csv")
data.head()
```

```python
data.isnull().mean()
```

```python
X = data.drop("bugs", axis=1)
y = data["bugs"]
```

#2. 분류 모델링

```python
y.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()

from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier()

from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier()

model.fit(X, y)
```

```python
from sklearn.metrics import classification_report
print( classification_report(y, model.predict(X)))
```

#3. Test셋에 대한 예측

```python
test = pd.read_csv("defect_test.csv")
```

```python
X_test = test.drop("bugs", axis=1)
y_test = test["bugs"]
```

```python
print( classification_report(y_test, model.predict(X_test)))
```

```python
for i in range(1,100):
  print("Start")
  print( i, "th record, prediction:", model.predict( X_test[i-1:i]))
  print( i, "th record, actual value:", y_test[i-1])
  print("End")
```
