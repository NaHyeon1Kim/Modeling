```python
import os #운영체제 라이브러리
import pandas as pd #데이터 분석 라이브러리

files1 = os.listdir('103')
files2 = os.listdir('105')
```

```python
x = []
y = []

def add_file(path):
  with open(path, "r", encoding='utf-8') as f:
    s = f.read()
    x.append(s)
```

```python
for f in files1:
  add_file("103/"+f)
  y.append(0)

for f in files2:
  add_file("105/"+f)
  y.append(1)
  #103폴더와 105폴더 범주를 각각 0과 1로 나눔
```

```python
len(y)
```

```python
df = pd.DataFrame( { 'label':y, 'articles':x } )
```

3. 토크나이저 만들기 (명사 추출)

```python
!pip install konlpy
from konlpy.tag import Okt
from konlpy.tag import Kkma

morphs = Okt()
def token(text):
  tokens_ko = morphs.nouns(text)
  return tokens_ko

morphs_k = Kkma()
def token_k(text):
  tokens_ko = morphs_k.nouns(text)
  return tokens_ko

token( df['articles'][0] )
```

CountVectorizer

- 단어의 빈도만 기반으로 계산해서 모든 단어 동일 가중치

TfidfVectorizer
